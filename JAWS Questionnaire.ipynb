```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
an1 = df[['What is your gender?', 'Sum']]
an1.columns = ['t', 'value']
an1
mod = ols('value ~ t',
                data=an1).fit()
                
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

```python
import pingouin as pg

aov = pg.anova(data=an1, dv='value', between='t', detailed=True)
print(aov)
```
